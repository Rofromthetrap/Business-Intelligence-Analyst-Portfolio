```python
td = TDClient(apikey=api_key)
ts = td.time_series(
    symbol="BTC/USD",
    exchange="Binance",
    interval="1day",
    outputsize=30,
    timezone="Africa/Dar_es_Salaam",
)
```

```python
# Returns pandas.DataFrame
df = ts.as_pandas()
df
```

```python


# Create a line chart
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['close'], linewidth=2, color='#1f77b4', label='Close Price')

# Customize the chart
plt.title('BTC/USD Price - Last 30 Days', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Display the chart
plt.show()
```
